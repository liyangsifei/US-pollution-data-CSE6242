```python
import pandas as pd
import numpy as np
```

### 1. CO

```python
# CO data from 2017-2020
keep_columns=["Latitude","Longitude","State Name","County Name","City Name","Date Local","Arithmetic Mean","1st Max Value","1st Max Hour","AQI"]

#2017
CO2017=pd.read_csv("CO_2017.csv")
CO2017_filter=CO2017[CO2017["Sample Duration"]=="8-HR RUN AVG END HOUR"].copy() # CO data keep all rows with 8-hour avg standard
CO2017_final=CO2017_filter[keep_columns]
print("There are {} observations of CO in 2017.".format(len(CO2017_final)))

#2018
CO2018=pd.read_csv("CO_2018.csv")
CO2018_filter=CO2018[CO2018["Sample Duration"]=="8-HR RUN AVG END HOUR"].copy()
CO2018_final=CO2018_filter[keep_columns]
print("There are {} observations of CO in 2018.".format(len(CO2018_final)))

#2019
CO2019=pd.read_csv("CO_2019.csv")
CO2019_filter=CO2019[CO2019["Sample Duration"]=="8-HR RUN AVG END HOUR"].copy()
CO2019_final=CO2019_filter[keep_columns]
print("There are {} observations of CO in 2019.".format(len(CO2019_final)))

#2020
CO2020=pd.read_csv("CO_2020.csv")
CO2020_filter=CO2020[CO2020["Sample Duration"]=="8-HR RUN AVG END HOUR"].copy()
CO2020_final=CO2020_filter[keep_columns]
print("There are {} observations of CO in 2020.".format(len(CO2020_final)))
CO2020_final.head()
```

```python
# merge four data frames vertically and re-order rows
CO_2017_2020=pd.concat([CO2017_final,CO2018_final,CO2019_final,CO2020_final], ignore_index=True)
CO_2017_2020_sort=CO_2017_2020.sort_values(by=["State Name","County Name","City Name","Date Local"])

# calculate avg of two site per county and reset index
CO_2017_2020_sort2=CO_2017_2020_sort.groupby(["State Name","County Name","City Name","Date Local"]).mean().reset_index()

#coordinates
CO_coordinate=CO_2017_2020_sort2[["State Name","County Name","City Name","Latitude","Longitude"]]

# rename and drop columns
CO_2017_2020_sort2=CO_2017_2020_sort2.drop(columns=["Latitude","Longitude"])
CO_2017_2020_sort2=CO_2017_2020_sort2.rename(columns={"State Name":"State","County Name":"County","City Name":"City", "Arithmetic Mean":"CO Mean","1st Max Value":"CO 1st Max Value","1st Max Hour":"CO 1st Max Hour","AQI":"CO AQI"})

print("There are total {} observations of CO from 2017-2020.".format(len(CO_2017_2020_sort2)))
CO_2017_2020_sort2.head()
```

### Merge all four data together

```python
CO_SO2=pd.merge(CO_2017_2020_sort2, SO_2017_2020_sort2, how="outer", on=["State","County","City","Date Local"])
O3_NO2=pd.merge(O_2017_2020_sort2, NO_2017_2020_sort2, how="outer", on=["State","County","City","Date Local"])
AllFour=pd.merge(CO_SO2, O3_NO2, how="outer", on=["State","County","City","Date Local"])
print("There are total {} observations of all from 2017-2020.".format(len(AllFour)))
AllFour.head()
```
